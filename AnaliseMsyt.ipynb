1. Distribuição dos inscritos (em milhões):

```python


# Carregar o arquivo CSV em um DataFrame
msyt = pd.read_csv('MostSubYouTube.csv', encoding='latin-1')

# Histograma dos inscritos
plt.hist(msyt['Subscribers (millions)'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Inscritos (milhões)')
plt.ylabel('Número de Canais')
plt.title('Distribuição dos Inscritos')
plt.show()

# Média e mediana dos inscritos
media_inscritos = msyt['Subscribers (millions)'].mean()
mediana_inscritos = msyt['Subscribers (millions)'].median()
print(f'Média dos inscritos: {media_inscritos:.2f} milhões')
print(f'Mediana dos inscritos: {mediana_inscritos:.2f} milhões')
```

2. Distribuição por idioma primário:

```python
# Contagem dos canais por idioma primário
idioma_counts = msyt['Primary language'].value_counts()

# Gráfico de barras
idioma_counts.plot(kind='bar', color='lightgreen')
plt.xlabel('Idioma Primário')
plt.ylabel('Número de Canais')
plt.title('Distribuição por Idioma Primário')
plt.xticks(rotation=45, ha='right')
plt.show()
```

3. Categorias de canais mais populares:

```python
# Contagem dos canais por categoria
categoria_counts = msyt['Category'].value_counts()

# Gráfico de barras
categoria_counts.plot(kind='bar', color=['orange', 'lightblue', 'lightgreen', 'pink', 'yellow'])
plt.xlabel('Categoria')
plt.ylabel('Número de Canais')
plt.title('Distribuição por Categoria')
plt.xticks(rotation=45)
plt.show()
```

4. Países com mais canais populares:

```python
# Contagem dos canais por país
pais_counts = msyt['Country'].value_counts().head(10)  # Exibindo os 10 países com mais canais

# Gráfico de barras horizontais
pais_counts.plot(kind='barh', color='salmon')
plt.xlabel('Número de Canais')
plt.ylabel('País')
plt.title('Top 10 Países com Mais Canais Populares')
plt.show()
```

5. Marcas vs. Canais Individuais:

```python
# 5. Marcas vs. Canais Individuais:
tipo_counts = msyt['Brand channel'].value_counts()

# Gráfico de barras
plt.bar(tipo_counts.index, tipo_counts.values, color=['lightcoral', 'lightskyblue'])
plt.xlabel('Tipo de Canal')
plt.ylabel('Número de Canais')
plt.title('Proporção de Canais de Marca e Individuais')
plt.show()

```

6. Distribuição dos inscritos por país:

```python
# Plotar a média de inscritos por país (apenas os 6 primeiros)
plt.figure(figsize=(10, 6))
pais_media_inscritos = msyt.groupby('Country')['Subscribers (millions)'].mean().sort_values(ascending=False)
pais_media_inscritos_top6 = pais_media_inscritos.head(6)
plt.bar(pais_media_inscritos_top6.index, pais_media_inscritos_top6.values, color='salmon')
plt.xlabel('País')
plt.ylabel('Média de Inscritos (milhões)')
plt.title('Média de Inscritos por País (Top 6)')
plt.xticks(rotation=45)
plt.show()
```

7. Idioma primário dos canais mais populares por categoria:

```python
# Contar o número de canais por categoria e idioma primário
idioma_categoria_counts = msyt.groupby('Category')['Primary language'].value_counts().unstack().fillna(0)

# Escolher os N idiomas mais comuns em cada categoria
N = 5  # Definir o número de idiomas a serem mostrados
top_idiomas_por_categoria = idioma_categoria_counts.apply(lambda x: x.nlargest(N), axis=1)

# Gráfico de barras agrupadas
top_idiomas_por_categoria.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Categoria')
plt.ylabel('Número de Canais')
plt.title('Idioma Primário dos Canais por Categoria (Top 5 Idiomas)')
plt.xticks(rotation=45)
plt.legend(title='Idioma Primário')
plt.show()
```

8. Distribuição dos inscritos por categoria:

```python
plt.figure(figsize=(10, 6))
categoria_media_inscritos = msyt.groupby('Category')['Subscribers (millions)'].mean().sort_values(ascending=False)
plt.bar(categoria_media_inscritos.index, categoria_media_inscritos.values, color='lightblue')
plt.xlabel('Categoria')
plt.ylabel('Média de Inscritos (milhões)')
plt.title('Média de Inscritos por Categoria')
plt.xticks(rotation=45)
plt.show()
```

9. Distribuição dos inscritos por idioma primário:

```python
plt.figure(figsize=(10, 6))
idioma_media_inscritos = msyt.groupby('Primary language')['Subscribers (millions)'].mean().sort_values(ascending=False)
plt.bar(idioma_media_inscritos.index, idioma_media_inscritos.values, color='lightgreen')
plt.xlabel('Idioma Primário')
plt.ylabel('Média de Inscritos (milhões)')
plt.title('Média de Inscritos por Idioma Primário')
plt.xticks(rotation=45)
plt.show()
```

10. Comparação entre inscritos de canais de marca e individuais:

```python
plt.figure(figsize=(6, 6))
inscritos_por_tipo = msyt.groupby('Brand channel')['Subscribers (millions)'].mean()
plt.bar(inscritos_por_tipo.index, inscritos_por_tipo.values, color=['lightcoral', 'lightskyblue'])
plt.xlabel('Tipo de Canal')
plt.ylabel('Média de Inscritos (milhões)')
plt.title('Comparação entre Inscritos de Canais de Marca e Individuais')
plt.show()
```
