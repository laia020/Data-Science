Neste notebook, analisamos o PIB (GDP - a soma da produção econômica total de um país em um determinado período) de alguns dos maiores países do mundo ao longo dos anos. Os países selecionados para esta análise são Brasil, Estados Unidos, China, Índia e Alemanha. Os dados utilizados são fictícios e foram criados para simular dados reais do World Bank Open Data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Os dados foram criados diretamente no notebook, simulando o PIB (GDP) em dólares correntes de 2010 a 2019 para os países selecionados.

```python
# Criar dados fictícios para a análise
data = {
    "Country Name": ["Brazil", "United States", "China", "India", "Germany"] * 10,
    "Year": [i for i in range(2010, 2020) for _ in range(5)],
    "GDP": [
        2.1e12, 15.0e12, 6.1e12, 1.7e12, 3.4e12,
        2.3e12, 15.5e12, 7.0e12, 1.8e12, 3.5e12,
        2.4e12, 16.2e12, 8.0e12, 1.9e12, 3.6e12,
        2.5e12, 16.8e12, 9.0e12, 2.0e12, 3.7e12,
        2.6e12, 17.5e12, 10.2e12, 2.2e12, 3.8e12,
        2.8e12, 18.1e12, 11.3e12, 2.3e12, 4.0e12,
        2.9e12, 18.8e12, 12.6e12, 2.5e12, 4.2e12,
        3.0e12, 19.4e12, 14.1e12, 2.7e12, 4.4e12,
        3.1e12, 20.2e12, 15.5e12, 2.9e12, 4.6e12,
        3.2e12, 21.0e12, 17.2e12, 3.1e12, 4.8e12
    ]
}
```

```python
# Criar o DataFrame
df = pd.DataFrame(data)

# Visualizar os primeiros dados
df.head()
```

Gráfico de Linha: Mostra a evolução do PIB ao longo dos anos.

```python
# Gráfico de Linha
plt.figure(figsize=(10, 6))
for country in df["Country Name"].unique():
    subset = df[df["Country Name"] == country]
    plt.plot(subset["Year"], subset["GDP"], label=country)
plt.xlabel("Year")
plt.ylabel("GDP (current US$)")
plt.title("GDP Over Time")
plt.legend()
plt.show()
```

Gráfico de Barra: Mostra o PIB dos países no ano mais recente disponível.

```python
# Gráfico de Barra
plt.figure(figsize=(10, 6))
latest_year = df["Year"].max()
subset = df[df["Year"] == latest_year]
plt.bar(subset["Country Name"], subset["GDP"])
plt.xlabel("Country")
plt.ylabel("GDP (current US$)")
plt.title(f"GDP in {latest_year}")
plt.show()
```

Histograma: Mostra a distribuição dos valores do PIB.

```python
# Histograma
plt.figure(figsize=(10, 6))
plt.hist(df["GDP"], bins=10, edgecolor="black")
plt.xlabel("GDP (current US$)")
plt.ylabel("Frequency")
plt.title("Distribution of GDP Values")
plt.show()
```

Gráfico de Pizza: Mostra a distribuição percentual do PIB no ano mais recente disponível.

```python
# Gráfico de Pizza
plt.figure(figsize=(8, 8))
plt.pie(subset["GDP"], labels=subset["Country Name"], autopct="%1.1f%%", startangle=140)
plt.title(f"GDP Distribution in {latest_year}")
plt.show()
```

Scatter Plot: Mostra a evolução do PIB ao longo dos anos de forma dispersa.

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
for country in df["Country Name"].unique():
    subset = df[df["Country Name"] == country]
    plt.scatter(subset["Year"], subset["GDP"], label=country)
plt.xlabel("Year")
plt.ylabel("GDP (current US$)")
plt.title("GDP Over Time (Scatter)")
plt.legend()
plt.show()
```

A análise gráfica permite observar o crescimento econômico dos países selecionados e comparar seus desempenhos ao longo do tempo.
